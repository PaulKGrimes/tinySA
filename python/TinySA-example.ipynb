# Sample notebook to interact with NanoVNA

```python
%matplotlib inline
from tinysa import TinySA
#reload script
# %run nanovna.py
```

```python
# create instance and connect to the device
sa = TinySA()
```

```python
sa.set_sweep(1e6, 300e6)
sa.fetch_frequencies()
```

```python
s11 = sa.data(1)
```

```python
print(sa.unit)
```

```python
sa.linmag(s11)
```

```python
sa.send_command("info")
sa.fetch_buffer()
```
